## Inheritance

```python
#parent class
class vehicle:
    def start(self):
        print("Car Is Started...")
#child class
class car(vehicle):
    def music(self):
        print("Music Is On ...")
c1=car()
c1.start()
c1.music()
```

```python
class person():
    def walk(self):
        print("StarT Walking..")
class student(person):
    def study(self):
        print("Start Study..")
s1=student()
s1.walk()
s1.study()
```

##### Vehicle System

```python
class vehicle:
    def start(self):
        print("Vehicle Is Started")
    def stop(self):
        print("Vehicle Is Stopped")
class car(vehicle):
    def ac(self):
        print("AC Is Turned On")
c=car()
c.start()
c.stop()
c.ac()
```

```python
class father:
    def voice(self):
        print("Voice Is High Pitch And Mascuiline Voice")
    def height(self):
        print("Height Is 6 Feet")
    def jawline(self):
        print("Sharp Jawline")
class son(father):
    def skin_color(self):
        print("Skin Color Is White")
    def nose(self):
        print("Sharp Nose")
s=son()
s.skin_color()
s.voice()
s.nose()
s.jawline()
        
```

```python
class Animal:
    def eat(self):
        print("Animal Eat Different Food")
class Dog(Animal):
    def sleep(self):
        print('Different Animals Are Sleeping')
class Puppy(Dog):
    def bark(self):
        print("Woof Woof Q!")
puppy=Puppy()
puppy.eat()
puppy.sleep()
puppy.bark()
```

### Hierarchical Inheritance

```python
class Vehicle:
    def start(self):
        print("Vehicle Is Started")
class Car(Vehicle):
    def stop(self):
        print("Car Is Stopped")
class Bike(Vehicle):
    def helmet(self):
        print("Wear Helmet")
car=Car()
car.start()
car.stop()
bike=Bike()
bike.start()
bike.helmet()

```
