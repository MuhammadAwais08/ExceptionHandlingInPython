#### OOP(Object Oriented Programming) 
**Class=Blueprint**
*Object=Real Thing Created from Class*
*Attributes=Properties Of Object*

```python
class Car:
    pass
car1=Car()
car1.brand="Rolls Royce"
car1.color="White"
car1.model="Dawn"
print(car1.brand)
print(car1.color)
print(car1.model)
```

```python
class Mobile:
    #pass
    m1=Mobile()
    m1.company="Iphone"
    m1.model="15 Pro Max"
    m1.RAM="16GB"
    m1.storage="256GB"
    print(m1.company)
    print(m1.model)
    print(m1.RAM)
    print(m1.storage)
```

##### Car Showroom

📖 Description

3 Car objects banao.

Har car ka:

Brand
Color
Price

Print karo.

Concept: Ek class, multiple objects.

```python
class carshowroom:
    pass
c1=carshowroom()
c1.color="White"
c1.brand="Lamborghini"
c1.price="100M$"
print("Object 1")
print(c1.color)
print(c1.brand)
print(c1.price)
c2=carshowroom()
c2.color="Black"
c2.brand="Bugatti"
c2.price="150M$"
print("Object 2")
print(c2.color)
print(c2.brand)
print(c2.price)
c3=carshowroom()
c3.color="White"
c3.brand="Rolls Royce"
c3.price="190M$"
print("Object 3")
print(c3.color)
print(c3.brand)
print(c3.price)

```
