#### *ValueError*

```python
#valueerror,zerodivisionerror and finally
try:
    y=int(input("Enter ANy Number:"))
    print(y)
except ValueError:
    print("Enter Only Numbers")
finally:
    print("Program Is Finished")
```

#### *ZeroDivisionError*

```python
try:
    num1=int(input("Enter 1st Number:"))
    num2=int(input("Enter 2nd Number:"))
    d=num1/num2
    print("Division Of Two Numbers=",d)
except ZeroDivisionError:
    print("Number Is'nt Divisible By 0")
finally:
    print("Calculation Is Performed")
```

#### *Mini Calculator*

```python
try:
    X=int(input("Enter 1st NUmber:"))
    Y=int(input("Enter 2nd NUmber:"))
    ope=(input("Enter Operation You Want To Perform:"))
    if ope=="Add":
        Z=X+Y
        print(f"Sum Of Two Numbers= {Z}")
    elif ope=="Sub":
        Z=X-Y
        print("Difference Of Two Numbers=",Z)
    elif ope=="Multi":
        Z=X*Y
        print(f"Product OF Two Numbers = {Z}")
    elif ope=="Divi":
        Z=X/Y
        print(f"Division Of Two Numbers = {Z}")
    else:
        print("Invalid Operation")
except ValueError:
    print("Please Enter Integers")
except ZeroDivisionError:
    print("Number Is'nt Divisible By 0")
finally:
    print("Operation Perfrom Successfully")
```
